### Pandas can get data from different types of files.
Below we will see examples of the following files
-  `csv`
-   `xls` 
-    `JSON` 


```python
import pandas as pd 
```

### CSV

1. Create a variable receiving the file path to make the code easier to read
2. Create a variable dataframe  receving the pandas method `.read_csv()`
3. Atributs
   1. path of the csv archive
   2. sep (character separator)
   3. encoding (set the decoder list )
   4. decimal (character divisor of decimal houses)

```python
path_csv = '../datasets/03a_AcompanhamentoOrcamentario_2018JanDez_V3_20190517.csv'
df_csv = pd.read_csv(path_csv, sep=';', encoding='latin1', decimal=',')
df_csv.head()
```

### XLSX (Excel)

1. Create a variable receiving the file path to make the code easier to read

3. Atributs
